```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
```

```python
credit_data = pd.read_csv("/content/credit_risk_dataset.csv")
credit_data.head()
```

```python
print(credit_data.describe())
```

```python
credit_data[['person_age','person_emp_length']]
```

```python
import matplotlib.pyplot as plt
plt.bar(credit_data['person_age'],credit_data['person_emp_length'])
```

```python
credit_data = credit_data[
    (credit_data['person_emp_length'] <= credit_data['person_age'] - 20) &
    (credit_data['person_emp_length'] >= 0)
]
```

```python
credit_data = credit_data[
    (credit_data['person_age'] >= 20) &
    (credit_data['person_age'] <= 75)
]
credit_data.head()
```

```python
max_emp_len = credit_data['person_emp_length'].quantile(0.95)
min_emp_len = credit_data['person_emp_length'].quantile(0.20)
```

```python
credit_data= credit_data[credit_data['person_emp_length'] < max_emp_len]
```

```python
X =credit_data[['person_age', 'person_income',
       'person_emp_length', 'loan_amnt', 'loan_percent_income']]

y  = credit_data['loan_status']
```

```python
credit_data.isna().sum()
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=28)
```

```python
scaler=StandardScaler()
x_train_scaled =scaler.fit_transform(X_train)
x_test_scaled= scaler.transform(X_test)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train_scaled,y_train)
```

```python
y_pred= model.predict(x_test_scaled)
accuracy= accuracy_score(y_test,y_pred)
```

```python
print("Accuracy", accuracy)
print("classification report",classification_report(y_test,y_pred))
```

```python
model.score(x_test_scaled,y_test)
```
